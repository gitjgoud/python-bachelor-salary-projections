# Select Eng Degree Popularity vs Google Search Frequency

### Summary

This analysis looks at the correlation between four engineering disciplines and google search frequencies (search terms characterized by field of study by google). These degrees were selected because they several decades of NCES reporting history (most individual degrees do not have reports prior to 2013).

##### Preparing popularity data

- [Google trends](https://trends.google.com/trends/explore?date=all&geo=US&q=Chemical%20Engineering,Civil%20Engineering,Electrical%20engineering,Mechanical%20Engineering) was used to compare search frequency for the selected degrees. Google provided the option to filter the results by characterizing the search term by phrases like "Field of Study", "Discipline", and "Professional Field" but a single phrase was not available for all degrees - therefore no filter was applied.
- The bulk of data cleaning and formatting was performed using spreadsheets prior to exporting to csv. Per instruction this was not uploaded to repo but can be found [here](https://docs.google.com/spreadsheets/d/11Q0bDMa9DZ216V0jylm8nLQnoD6onql_xNSoIMpHkzM/edit?usp=sharing).

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
!cat eng_degree_pop_export.csv
```

```python
eng_degree_pop = pd.read_csv('eng_degree_pop_export.csv')
eng_degree_pop
```

```python
plt.figure(figsize =(20,5))
ax = plt.subplot(1,2,1)
xvalues=range(15)
ax.set_xticks(range(15))
ax.set_xticklabels(range(2005,2020))
plt.plot(xvalues,eng_degree_pop['Chem_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Civil_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Elec_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Mech_Eng_Rel'])
plt.legend(['ChemE','CivE','ElecE','MechE'],loc=0)
plt.xlabel('Year')
plt.ylabel('Relative Popularity')
plt.title('Relative Popularity by Degree')
plt.grid(b=True, which='major',axis='y')
plt.grid(b=True, which='major',axis='x')

ax = plt.subplot(1,2,2)
xvalues=range(15)
ax.set_xticks(range(15))
ax.set_xticklabels(range(2005,2020))
plt.plot(xvalues,eng_degree_pop['Chem_Eng_Rel_SearhFreq'])
plt.plot(xvalues,eng_degree_pop['Civil_Eng_Rel_SearhFreq'])
plt.plot(xvalues,eng_degree_pop['Elec_Eng_Rel_SearhFreq'])
plt.plot(xvalues,eng_degree_pop['Mech_Eng_Rel_SearhFreq'])
plt.legend(['ChemE','CivE','ElecE','MechE'],loc=0)
plt.xlabel('Year')
plt.ylabel('Relative Search Frequency')
plt.title('Google Search Interest by Degree')
plt.grid(b=True, which='major',axis='y')
plt.grid(b=True, which='major',axis='x')
plt.plot()
```
