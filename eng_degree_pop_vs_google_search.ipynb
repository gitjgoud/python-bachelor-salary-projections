```python
eng_degree_pop = pd.read_csv('eng_degree_pop_export.csv')
eng_degree_pop
```

```python
plt.figure(figsize =(20,5))
ax = plt.subplot(1,2,1)
xvalues=range(15)
ax.set_xticks(range(15))
ax.set_xticklabels(range(2005,2020))
plt.plot(xvalues,eng_degree_pop['Chem_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Civil_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Elec_Eng_Rel'])
plt.plot(xvalues,eng_degree_pop['Mech_Eng_Rel'])
plt.legend(['ChemE','CivE','ElecE','MechE'],loc=0)
plt.xlabel('Year')
plt.ylabel('Relative Popularity')
plt.title('Relative Popularity by Degree')
plt.grid(b=True, which='major',axis='y')

ax = plt.subplot(1,2,2)
xvalues=range(15)
ax.set_xticks(range(15))
ax.set_xticklabels(range(2005,2020))
plt.plot(xvalues,eng_degree_pop['Chem_Eng_AnnAvg_SearchFreq'])
plt.plot(xvalues,eng_degree_pop['Civil_Eng_AnnAvg_SearchFreq'])
plt.plot(xvalues,eng_degree_pop['Elec_Eng_AnnAvg_SearchFreq'])
plt.plot(xvalues,eng_degree_pop['Mech_Eng_AnnAvg_SearchFreq'])
plt.legend(['ChemE','CivE','ElecE','MechE'],loc=0)
plt.xlabel('Year')
plt.ylabel('Search Frequency')
plt.title('Google Search Interest by Degree')
plt.grid(b=True, which='major',axis='y')
plt.plot()
```
